## Restoring a model

After training a model, we can: 
1. Download a model from wandb
2. Instantiade the model and its sampler from the artifact

```python
problem_id = "beams2d"
seed = 1

use_wandb = os.getenv("USE_WANDB") == "True"

## Downloading a model from wandb
if use_wandb:
    # This is the name of the model in wandb, for the sake of the example we will use a model that is stored on the official engiopt wandb project. If you want to use a model that you trained yourself, you can change the artifact_path to the path of your model.
    artifact_path = f"engibench/engiopt/{problem_id}_cgan_cnn_2d_generator:seed_{seed}"

    api = wandb.Api()
    artifact = api.artifact(artifact_path, type="model")
    artifact_dir = artifact.download()
    ckpt_path = os.path.join(artifact_dir, "generator.pth")

    ckpt = th.load(ckpt_path, map_location=device)

    # We will also need the "run" object to get the model configuration which was used to train the model.
    run = artifact.logged_by()

    # This is the model definition that was used to train the model. You can find the definition in the engiopt/cgan_cnn_2d/cgan_cnn_2d.py file.
    model = Generator(
        latent_dim=run.config["latent_dim"], n_conds=len(problem.conditions), design_shape=problem.design_space.shape
    )

    model.load_state_dict(ckpt["generator"])

else:
    # If you don't want to use wandb, you can load the model from the local file system (if you've trained it yourself)
    ckpt = th.load("generator.pth", map_location=device)
    model = Generator(latent_dim=32, n_conds=len(problem.conditions), design_shape=problem.design_space.shape)
    model.load_state_dict(ckpt["generator"])

model.eval()
model.to(device)
```

## Sampling from a model

Now we can use the model to generate designs.

```python
n_samples = 3

conditions = [
    {"volfrac": 0.35, "rmin": 2.0, "forcedist": 0.2, "overhang_constraint": 0.0},
    {"volfrac": 0.45, "rmin": 2.0, "forcedist": 0.2, "overhang_constraint": 0.0},
    {"volfrac": 0.4, "rmin": 2.0, "forcedist": 0.35, "overhang_constraint": 0.0},
]

conditions_tensor = th.tensor([list(c.values()) for c in conditions], device=device)
conditions_tensor = conditions_tensor.unsqueeze(-1).unsqueeze(-1)  # Reshape to match the expected input shape for the model

# Sample noise as generator input
if use_wandb:
    z = th.randn((n_samples, run.config["latent_dim"], 1, 1), device=device, dtype=th.float)
else:
    z = th.randn((n_samples, 32, 1, 1), device=device, dtype=th.float)

# Generate a batch of designs
gen_designs = model(z, conditions_tensor)

gen_designs = gen_designs.detach().cpu().numpy()
gen_designs = gen_designs.squeeze()  # Removes the extra dimension
np.clip(
    gen_designs, 0, 1, out=gen_designs
)  # Sometimes our models generate designs outside the design space... this is a quick fix

print(gen_designs.shape)

# Render the designs
for i in range(n_samples):
    problem.render(gen_designs[i])
```

## Evaluating the model

### Optimizing generated designs

Let's first optimize the generated designs, to see how far away they are from optimal designs

```python
opt_designs = []
opt_histories = []
for i in range(n_samples):
    opt_design, opt_history = problem.optimize(
        starting_point=gen_designs[i],
        config=conditions[i],
    )
    opt_histories.append(opt_history)
    opt_designs.append(opt_design)
```

### Eyeball testing the model

Render the optimal designs next to the generated designs to get a visual idea

```python
for i in range(n_samples):
    fig_opt, ax_opt = problem.render(opt_designs[i])
    fig_gen, ax_gen = problem.render(gen_designs[i])
    ax_opt.set_title("Optimal")
    ax_gen.set_title("Generated")
    plt.show()
```

### Qualitative evaluation

Now we can also use metrics to get a quantitative idea of how good the generated designs are. Here we compute optimality gaps between the generated designs and optimized designs in terms of objective values (structural compliance for beams). We have other metrics available, see the [metrics.py](https://github.com/IDEALLab/EngiOpt/blob/main/engiopt/metrics.py) file for more details.

```python
objective_values_opt = [problem.simulate(opt_designs[i], config=conditions[i]) for i in range(n_samples)]
optimality_gaps = [metrics.optimality_gap(opt_histories[i], baseline=objective_values_opt[i]) for i in range(n_samples)]

iog = [optimality_gaps[i][0][0] for i in range(n_samples)]
cog = [np.sum(optimality_gaps[i]) for i in range(n_samples)]
fog = [optimality_gaps[i][-1][0] for i in range(n_samples)]

print("\033[96m" + f"Initial optimality gaps: {iog}" + "\033[0m")
print("\033[96m" + f"Cumulative optimality gaps: {cog}" + "\033[0m")
print("\033[96m" + f"Final optimality gaps: {fog}" + "\033[0m")
```

### But what is an optimality gap?

Let's compare the design generated by the GAN with a design generated by uniformly filling the design space with the desired volume fraction, showing that the GAN provides a better starting point for the optimization.

```python
# choosing one design that was generated by the GAN
design_nb = 0
design = gen_designs[design_nb]
opt_history = opt_histories[design_nb]

conditions_nb = conditions[design_nb]  # these are the conditions that were used to generate the design
volfrac = conditions_nb["volfrac"]

# let's generate a design with the desired volume fraction
design_shape = problem.design_space.shape
# generate the design with shape design_shape uniformly filled with the desired volume fraction
design = np.ones(design_shape) * volfrac

# let's optimize the design
opt_design_uniform, opt_history_uniform = problem.optimize(design, config=conditions_nb)

obj_values_uniform = [optistep.obj_values[0] for optistep in opt_history_uniform]
obj_values_gen = [optistep.obj_values[0] for optistep in opt_history]

# let's plot both optimization histories
plt.plot(obj_values_uniform, label="Uniformly filled design")
plt.plot(obj_values_gen, label="GAN design")
plt.xlabel("Optimization iteration")
plt.ylabel("Objective value")
plt.yscale("log")
plt.title("Optimization history")
plt.legend()
plt.show()
```

This highlights that GAN-generated designs provide strong initializations for optimization, being significantly closer to the optimal solution than uniform starting points. Interestingly, however, the optimizer can sometimes find a better final design when starting from a uniform initialization. This is because gradient-based solvers may become trapped in local optima. Thus, while inverse design can accelerate convergence, it does not guarantee avoidance of suboptimal local minima—and poor initializations may not be recoverable at all.

The optimality gap measures how far a design’s objective value is from the optimum; smaller gaps indicate more promising starting points.

```python
# finally let's have a look at the optimized designs
opt_gans = opt_designs[design_nb]

fig, ax = problem.render(opt_gans)
fig2, ax2 = problem.render(opt_design_uniform)
ax.set_title("Optimized design starting from GAN")
ax2.set_title("Optimized design starting from uniformly filled design")

plt.show()
```
